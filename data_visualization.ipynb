## Line Graph

```python
x =df['Timestamp']
y = df['LecturerStress']
plt.plot(x, y,label="line1")
plt.plot(x, y,'black')
plt.title('TimestampVsLecturerStress')
plt.xlabel('Timestamp')
plt.ylabel('LecturerStress')
plt.show()
```

```python
fig = plt.figure(figsize =(10, 7))
x =df['Timestamp']
y = df['LecturerStress']
z =df['FamilyStress']
a=df['AssignmentStress']
plt.plot(x, y,'black')
plt.plot(x, z,'red')
plt.plot(x, a,'-.')
plt.legend(["LecturerStress", "FamilyStress","AssignmentStress"], loc ="upper right")
plt.title('StressComparision')
plt.xlabel('Timestamp')
plt.ylabel('Stress')
```
